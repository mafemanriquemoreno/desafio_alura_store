### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

#1. Análisis de facturación



```python
diccionario_tiendas = {  "Tienda 1": tienda,  "Tienda 2": tienda2,  "Tienda 3": tienda3,     "Tienda 4": tienda4 }

for nombre_tienda, datos_tienda in diccionario_tiendas.items():

    #Calculamos el ingreso sumando la columna 'Precio'
    ingreso_total = datos_tienda['Precio'].sum()
    print(f"El ingreso total de la {nombre_tienda} es: ${ingreso_total:,.2f}")
```

# 2. Ventas por categoría

```python
print("Análisis de Ventas por Categoría")

for nombre_tienda, datos_tienda in diccionario_tiendas.items():

    print(f"\n--- {nombre_tienda} ---")
    # Calculamos el conteo de ventas para cada categoría
    ventas_por_categoria = datos_tienda['Categoría del Producto'].value_counts()

    # Resultado
    print(ventas_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
for nombre_tienda, datos_tienda in diccionario_tiendas.items():

    # Usare la columna 'Calificación' y se calculara su promedio con .mean()
    calificacion_promedio = datos_tienda['Calificación'].mean()
    print(f"La calificación promedio de la {nombre_tienda} es: {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
for nombre_tienda, datos_tienda in diccionario_tiendas.items():

    print(f"\n\n--- {nombre_tienda} ---")

    # Se cuenta la frecuencia de venta de cada producto
    conteo_productos = datos_tienda['Producto'].value_counts()

    # Se mostrara los 5 productos más vendidos usando .head()
    print("  ✅ Productos MÁS vendidos:")
    print(conteo_productos.head(5))

    # Se mostrara los 5 productos menos vendidos usando .tail()
    print("\n  ❌ Productos MENOS vendidos:")
    print(conteo_productos.tail(5))
```

# 5. Envío promedio por tienda

```python
for nombre_tienda, datos_tienda in diccionario_tiendas.items():

    # 1. Se selecciona la columna 'Costo de envío' y calculamos su promedio
    envio_promedio = datos_tienda['Costo de envío'].mean()

    print(f"Promedio costos de envio de la {nombre_tienda} es: ${envio_promedio:,.2f}")
```

## ** Análisis de Métricas **

```python
import matplotlib.pyplot as plt

ingresos = []
nombres_tiendas = list(diccionario_tiendas.keys())

for nombre in nombres_tiendas:
    ingreso = diccionario_tiendas[nombre]['Precio'].sum()
    ingresos.append(ingreso)

# Creamos grafico
plt.figure(figsize=(10, 6)) # Definicion del tamaño del gráfico (ancho, alto en pulgadas)

# Creamos el gráfico de barras. Eje X: nombres_tiendas, Eje Y: ingresos
plt.bar(nombres_tiendas, ingresos, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# Títulos y Etiquetas
plt.title('Ingresos Totales por Tienda', fontsize=16)

# Nombramos para el eje X y al eje Y
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Ingresos (en cientos de millones)', fontsize=12)

# Rotamos las etiquetas del eje X para que no se superpongan (opcional pero recomendado)
plt.xticks(rotation=45)

# Añadimos una cuadrícula para facilitar la lectura
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

# *Grafico de promedio *

```python
import matplotlib.pyplot as plt

# Calificaciones del promedio para cada tienda
calificaciones = []
nombres_tiendas = list(diccionario_tiendas.keys())

for nombre in nombres_tiendas:
    calificacion_media = diccionario_tiendas[nombre]['Calificación'].mean()
    calificaciones.append(calificacion_media)

# Creacion del grsfico
plt.figure(figsize=(10, 6))

plt.barh(nombres_tiendas, calificaciones, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# --- 3. Añadir Títulos y Etiquetas ---
plt.title('Calificación Promedio por Tienda', fontsize=16)
plt.xlabel('Calificación Promedio (de 1 a 5 estrellas)', fontsize=12)
plt.ylabel('Tiendas', fontsize=12)

plt.xlim(1, 5)

# Añadimos una cuadrícula vertical para facilitar la lectura
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

# Grafico **Costo envio promedio **

```python
import matplotlib.pyplot as plt

# Calculo del costo de envío promedio por cada tienda
costos_envio = []
nombres_tiendas = list(diccionario_tiendas.keys())

for nombre in nombres_tiendas:
    costo_medio = diccionario_tiendas[nombre]['Costo de envío'].mean()
    costos_envio.append(costo_medio)

# creacion del grafico
plt.figure(figsize=(10, 6))
plt.bar(nombres_tiendas, costos_envio, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

plt.title('Costo de Envío Promedio por Tienda', fontsize=16)
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Costo de Envío Promedio ($)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
